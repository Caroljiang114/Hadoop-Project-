```
path = "dbfs:/FileStore/tables/StudentGrades.csv"
 
grades_df = spark.read.load(path,
                    format='com.databricks.spark.csv', 
                    header='true',
                    inferSchema='true')
 
display(grades_df)
```

```
path = "dbfs:/FileStore/tables/testresult.csv"
 
testresult_df = spark.read.load(path,
                    format='com.databricks.spark.csv', 
                    header='true',
                    inferSchema='true')
 
display(testresult_df)
```

```
result_df = testresult_df.join(
    grades_df,
    (testresult_df["sid"] == grades_df["sid"]) & (testresult_df["examname"] == grades_df["examname"]),  
    "inner"  
).select(
    testresult_df["timestamp"],
    testresult_df["data"],
    testresult_df["type"],
    testresult_df["sid"],
    testresult_df["examname"],
    grades_df["result"]
).filter(
    (testresult_df["type"] == "TEMP") & (testresult_df["examname"] == "Midterm 2") 
)

result_df.show()
```

```
table_df=spark.table("temp_data_in_midterm2")
df = table_df.toPandas()
```

```
df.head(1)
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def classify_result(result):
    if result < 60:
        return 'Low Grade'
    elif result >= 60 and result < 80:
        return 'Normal Grade'
    elif result >= 80 :
        return 'High Grade'
    else:
        return 'Unknown'

df['result_group'] = df['result'].apply(classify_result)

df= df[['timestamp', 'data', 'type', 'sid','examname','result_group']]

print(df)

print(df.dtypes)

g = sns.FacetGrid(df, row="result_group", hue="sid", height=4, aspect=2)
g.map(sns.lineplot, "timestamp", "data")

g.set_titles("Result Group: {row_name}")

g.add_legend()

plt.show()

df.shape
```
